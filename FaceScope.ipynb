```python
from datetime import datetime
from dateutil.parser import parse

def get_zodiac_sign(birth_date_str):
    """
    Determine zodiac sign from birth date
    """
    # Parse the birth date string
    birth_date = parse(birth_date_str)
    month = birth_date.month
    day = birth_date.day

    # Zodiac date ranges
    zodiac_ranges = [
        (1, 20, 'Capricorn'), (2, 19, 'Aquarius'), (3, 21, 'Pisces'),
        (4, 20, 'Aries'), (5, 21, 'Taurus'), (6, 21, 'Gemini'),
        (7, 23, 'Cancer'), (8, 23, 'Leo'), (9, 23, 'Virgo'),
        (10, 23, 'Libra'), (11, 22, 'Scorpio'), (12, 22, 'Sagittarius'),
        (12, 31, 'Capricorn')  # Wrap around for Dec 22-31
    ]

    for end_day, end_month, sign in zodiac_ranges:
        if month == end_month:
            if day <= end_day:
                return sign
            elif month == 12 and day > 21:  # Special case for Capricorn
                return 'Capricorn'

    # Handle the transition between signs
    if (month == 1 and day > 19) or (month == 2 and day <= 18):
        return 'Aquarius'
    elif (month == 2 and day > 18) or (month == 3 and day <= 20):
        return 'Pisces'
    elif (month == 3 and day > 20) or (month == 4 and day <= 19):
        return 'Aries'
    elif (month == 4 and day > 19) or (month == 5 and day <= 20):
        return 'Taurus'
    elif (month == 5 and day > 20) or (month == 6 and day <= 21):
        return 'Gemini'
    elif (month == 6 and day > 21) or (month == 7 and day <= 22):
        return 'Cancer'
    elif (month == 7 and day > 22) or (month == 8 and day <= 22):
        return 'Leo'
    elif (month == 8 and day > 22) or (month == 9 and day <= 22):
        return 'Virgo'
    elif (month == 9 and day > 22) or (month == 10 and day <= 23):
        return 'Libra'
    elif (month == 10 and day > 23) or (month == 11 and day <= 22):
        return 'Scorpio'
    elif (month == 11 and day > 22) or (month == 12 and day <= 21):
        return 'Sagittarius'
    else:  # December 22 - January 19
        return 'Capricorn'
```

```python
import requests
import os
from PIL import Image
import urllib.parse
import time

def download_celebrity_images(csv_file_path):
    """
    Download celebrity images and organize by zodiac sign
    """
    # Create directory structure
    base_dir = "zodiac_dataset"
    zodiac_signs = [
        'Aries', 'Taurus', 'Gemini', 'Cancer', 'Leo', 'Virgo',
        'Libra', 'Scorpio', 'Sagittarius', 'Capricorn', 'Aquarius', 'Pisces'
    ]

    for sign in zodiac_signs:
        os.makedirs(os.path.join(base_dir, sign), exist_ok=True)

    import pandas as pd
    df = pd.read_csv(csv_file_path)

    for index, row in df.iterrows():
        name = row['Name']
        birth_date = row['Birth_Date']
        image_url = row['Image_URL']

        # Calculate zodiac sign
        zodiac_sign = get_zodiac_sign(birth_date)

        # Download image
        try:
            response = requests.get(image_url, stream=True)
            response.raise_for_status()

            # Create safe filename
            safe_name = "".join(c for c in name if c.isalnum() or c in (' ', '-', '_')).rstrip()
            filename = f"{safe_name.replace(' ', '_')}_{index}.jpg"
            filepath = os.path.join(base_dir, zodiac_sign, filename)

            # Save image
            with open(filepath, 'wb') as f:
                for chunk in response.iter_content(chunk_size=8192):
                    f.write(chunk)

            print(f"Downloaded: {name} -> {zodiac_sign}")

            # Add delay to avoid overwhelming servers
            time.sleep(0.5)

        except Exception as e:
            print(f"Failed to download {name}: {str(e)}")

# Usage
download_celebrity_images('celebrities.csv')
```

```python
import cv2
import os
from PIL import Image

def extract_faces_from_dataset(dataset_path):
    """
    Extract faces from downloaded celebrity images and save cropped faces
    """
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

    # Process each zodiac folder
    for zodiac_folder in os.listdir(dataset_path):
        folder_path = os.path.join(dataset_path, zodiac_folder)

        if os.path.isdir(folder_path):
            print(f"Processing {zodiac_folder}...")

            # Create a subfolder for cropped faces
            faces_folder = os.path.join(folder_path, 'cropped_faces')
            os.makedirs(faces_folder, exist_ok=True)

            for filename in os.listdir(folder_path):
                if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
                    image_path = os.path.join(folder_path, filename)

                    # Load image
                    img = cv2.imread(image_path)
                    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

                    # Detect faces
                    faces = face_cascade.detectMultiScale(gray, 1.1, 5)

                    # Crop and save each face
                    for i, (x, y, w, h) in enumerate(faces):
                        # Add padding around face
                        padding = int(min(w, h) * 0.1)  # 10% padding
                        x_start = max(0, x - padding)
                        y_start = max(0, y - padding)
                        x_end = min(img.shape[1], x + w + padding)
                        y_end = min(img.shape[0], y + h + padding)

                        face_crop = img[y_start:y_end, x_start:x_end]

                        # Save cropped face
                        face_filename = f"{filename[:-4]}_face_{i}.jpg"
                        face_path = os.path.join(faces_folder, face_filename)
                        cv2.imwrite(face_path, face_crop)

                    print(f"  Processed {filename}: Found {len(faces)} face(s)")

# Usage
extract_faces_from_dataset('zodiac_dataset')
```

```python
import cv2
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import cosine_similarity
import os
from collections import Counter
import matplotlib.pyplot as plt

class ZodiacFaceMatcher:
    def __init__(self):
        self.face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
        self.zodiac_signs = [
            'Aries', 'Taurus', 'Gemini', 'Cancer', 'Leo', 'Virgo',
            'Libra', 'Scorpio', 'Sagittarius', 'Capricorn', 'Aquarius', 'Pisces'
        ]
        self.training_encodings = {}
        self.encodings_by_zodiac = {sign: [] for sign in self.zodiac_signs}

    def extract_face_encoding(self, image_path):
        """Extract face encoding from image"""
        img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        if img is None:
            return None

        faces = self.face_cascade.detectMultiScale(img, 1.1, 5)

        if len(faces) == 0:
            print(f"No face detected in: {image_path}")
            return None

        # Get the largest face (usually the main subject)
        (x, y, w, h) = max(faces, key=lambda f: f[2] * f[3])
        face_roi = img[y:y+h, x:x+w]

        # Resize to standard size
        face_resized = cv2.resize(face_roi, (150, 150))

        return face_resized

    def load_training_data(self, dataset_path):
        """Load all training images"""
        print("Loading training data...")

        for zodiac_sign in self.zodiac_signs:
            sign_folder = os.path.join(dataset_path, zodiac_sign)

            if os.path.exists(sign_folder):
                print(f"Processing {zodiac_sign}...")

                for filename in os.listdir(sign_folder):
                    if filename.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp')):
                        image_path = os.path.join(sign_folder, filename)

                        encoding = self.extract_face_encoding(image_path)
                        if encoding is not None:
                            image_id = f"{zodiac_sign}_{filename}"
                            self.training_encodings[image_id] = {
                                'encoding': encoding,
                                'zodiac': zodiac_sign
                            }
                            self.encodings_by_zodiac[zodiac_sign].append(encoding)

        print(f"Loaded {len(self.training_encodings)} face encodings")

    def calculate_face_similarity(self, face1, face2):
        """Calculate similarity between two face encodings"""
        face1_norm = face1.astype(np.float32)
        face2_norm = face2.astype(np.float32)

        correlation = np.corrcoef(face1_norm.flatten(), face2_norm.flatten())[0,1]

        if np.isnan(correlation):
            return 1.0 if np.array_equal(face1, face2) else 0.0

        return max(0, correlation)

    def predict_zodiac_from_face(self, query_image_path):
        """Predict zodiac sign for query image"""
        query_encoding = self.extract_face_encoding(query_image_path)

        if query_encoding is None:
            return {"error": "No face detected in query image"}

        # Calculate similarity with all training faces
        similarities = []

        for image_id, data in self.training_encodings.items():
            similarity = self.calculate_face_similarity(query_encoding, data['encoding'])
            similarities.append((similarity, data['zodiac']))

        # Sort by similarity (highest first)
        similarities.sort(key=lambda x: x[0], reverse=True)

        # Get top N matches and vote for zodiac sign
        top_matches = similarities[:10]

        # Count zodiac votes
        zodiac_votes = Counter([match[1] for match in top_matches])

        # Get the most common zodiac sign
        predicted_zodiac = zodiac_votes.most_common(1)[0][0]
        confidence = zodiac_votes[predicted_zodiac] / sum(zodiac_votes.values())

        # Calculate average similarity for the predicted zodiac
        zodiac_matches = [sim for sim, zod in top_matches if zod == predicted_zodiac]
        avg_similarity = np.mean(zodiac_matches) if zodiac_matches else 0

        return {
            'predicted_zodiac': predicted_zodiac,
            'confidence': confidence,
            'avg_similarity': avg_similarity,
            'top_matches': [(sim, zod) for sim, zod in top_matches[:5]],
            'vote_breakdown': dict(zodiac_votes)
        }

def main():
    # Initialize the matcher
    matcher = ZodiacFaceMatcher()

    # Load your dataset
    dataset_path = "zodiac_dataset"  # Update this path to your actual dataset location

    print("Loading dataset...")
    matcher.load_training_data(dataset_path)

    # Test with a new image
    # Corrected string formatting for the local path example
    # Note: This path is for a local Windows machine and will not work directly in Colab.
    # You need to upload the image or provide a path to an image within the Colab environment.
    test_image = '/content/652805ec3b1a42.07930442.jpg'  # Replace with actual test image path

    if os.path.exists(test_image):
        print(f"\nPredicting zodiac for: {test_image}")
        result = matcher.predict_zodiac_from_face(test_image)

        if 'error' in result:
            print(result['error'])
        else:
            print(f"\nPrediction Results:")
            print(f"Predicted Zodiac: {result['predicted_zodiac']}")
            print(f"Confidence: {result['confidence']:.3f}")
            print(f"Average Similarity: {result['avg_similarity']:.3f}")
            print(f"Vote Breakdown: {result['vote_breakdown']}")
            print(f"Top 5 Matches: {result['top_matches']}")
    else:
        print(f"Test image not found: {test_image}")
        print("Please provide a valid image path for testing")

if __name__ == "__main__":
    main()
```
